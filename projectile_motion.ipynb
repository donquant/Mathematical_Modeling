# Постановка

Тело массы $m$ вылетает из точки (0,0) со скоростью $v_0$ под углом $\alpha$ к оси $х$ и движется в постоянном поле силы тяжести

$$
  F_{gr} = −mg,
$$

испытывая силу трения

$$
   F_{fr} = -\beta ({\textbf{v}}/{v}) = -\beta(v)\Big(\textbf{e}_x \frac{v_x}{v} + \textbf{e}_z \frac{v_z}{v}\Big).
$$

### Уравнения движения тела.

$$
  \frac{dx}{dt} = v_x
$$

$$
  \frac{dx}{dz} = v_z
$$

$$
  m\frac{dv_x}{dt} = F_{fr_x}(v) = -\beta(v)v_x
$$

$$
  m\frac{dv_z}{dt} = -mg + F_{fr_z}(v) = -mg -\beta(v)v_z
$$

$$
  v = \sqrt{v_x^2 + v_z^2}
$$

### Начальные условия.

$$
  x(0) = 0
$$

$$
  z(0) = 0
$$

$$
  v_x(0) = v_{x_0} = v_0cos\alpha
$$

$$
  v_z(0) = v_{z_0} = v_0sin\alpha
$$

$$
  0 \leq \alpha \leq \frac{\pi}{2}
$$

# Решение

Для аналитического решения рассмотрим силу трения пропорциональную первой степени скорости $\beta(v) = \beta_0v \implies$

$$
  m\frac{dv_x}{dt} = -\beta_0v_x
$$

$$
  m\frac{dv_z}{dt} = -mg -\beta_0v_z
$$

Перепишем уравнения для скорости в виде

$$
  \frac{dv_x}{v_x} = -(\beta_0/m)dt
$$

$$
  \frac{dv_z}{dt} = -g (-\beta_0/m)v_z
$$

$$
  v_x(0) = v_0cos\alpha
$$

$$
  v_z(0) = v_0sin\alpha
$$

Решим уравнение для $v_x$

$$
  lnv_x = -(\beta_0/m)dt + lnC \implies v_x(t) = Ce^{-(\beta_0/m)t}
$$

Из начального условия $v_x(0) = C = v_0cos\alpha$ получим решение


<font color='blue' size=5pt>
$$
  v_x(t) = v_0cos\alpha e^{-(\beta_0/m)t}.
$$
</font>

Решим уравнение для $v_z$

$$
  \frac{dv_z}{g+(\beta_0/m)v_z} = -dt \implies \frac{d(g+(\beta_0/m)v_z)}{g+(\beta_0/m)v_z} = -(\beta_0/m)dt
$$

$$
  \implies ln(g+(\beta_0/m)v_z) = -(\beta_0/m)e +lnC \implies v_z = C\frac{m}{\beta_0}e^{-(\beta_0/m)t} - \frac{m}{\beta_0}g
$$

Найдём $C$ из начального условия

$$
  v_z(0) = C\frac{m}{\beta_0} - \frac{m}{\beta_0}g = v_0 sin\alpha \implies C = \frac{\beta_0}{m}v_0 sin\alpha + g
$$

В итоге получаем решение

<font color='blue' size=5pt>
$$
  v_z(t) = (v_0 sin\alpha + g\frac{m}{\beta_0})e^{-(\beta_0/m)t} - \frac{m}{\beta_0}g
$$    
</font>

Заметим, что в отсутствии силы трения $\beta_0 \rightarrow 0$ полученные решения примут вид

$$
  v_x(t) = v_0cos\alpha
$$

$$
  v_z(t) = v_0sin\alpha - gt
$$

Получим теперь решения для координаты $x$ из уравнений

$$
  \frac{dx}{dt} = v_x(t) = v_0cos\alpha e^{-(\beta_0/m)t}
$$

$$
  x(0) = C - \frac{m}{\beta_0}v_0cos\alpha e^{-(\beta_0/m)t}
$$

Из начального условия

$$
  x(0) = C - \frac{m}{\beta_0}v_0cos\alpha = x_0 \implies C = x_0 + \frac{m}{\beta_0}v_0cos\alpha \implies
$$


<font color='blue' size=5pt>
$$
  x(t) = x_0 + \frac{m}{\beta_0}v_0cos\alpha(1 - e^{-(\beta_0/m)t})
$$
</font>

При $\beta_0 \rightarrow 0$ это решение примет вид

$$
  x(t) = x_0 + \frac{m}{\beta_0}v_0cos\alpha(1 - (1 - \frac{\beta_0}{m}t) = x_0 + v_0cos\alpha t
$$


Получим теперь решения для координаты $z$ из уравнения

$$
  \frac{dz}{dt} = v_z(t) = (v_0sin\alpha + g\frac{m}{\beta_0})e^{-(\beta_0/m)t} - \frac{m}{\beta_0}g
$$

$$
  z(0) = z_0
$$

Его общее решение

$$
  \frac{dz}{dt} = v_z(t) = - \frac{m}{\beta_0}(v_0sin\alpha + g\frac{m}{\beta_0})e^{-(\beta_0/m)t} - \frac{m}{\beta_0}gt + C
$$

Из начального условия

$$
  z(0) = - \frac{m}{\beta_0}(v_0sin\alpha + g\frac{m}{\beta_0})e^{-(\beta_0/m)0} - \frac{m}{\beta_0}g0 + C = z_0 \implies C = z_0 + \frac{m}{\beta_0}(v_0sin\alpha + g\frac{m}{\beta_0})
$$

таким образом получаем решение

<font color='blue' size=5pt>
$$
  z(t) = z_0 - \frac{m}{\beta_0}gt + \frac{m}{\beta_0}(v_0sin\alpha + g\frac{m}{\beta_0})(1-e^{-(\beta_0/m)t})
$$
</font>

## Итоговые уравнения

<font color='green' size=5pt>    
$$
  v_x(t) = v_0cos\alpha e^{-(\beta_0/m)t}
$$

$$
  v_z(t) = (v_0 sin\alpha + g\frac{m}{\beta_0})e^{-(\beta_0/m)t} - \frac{m}{\beta_0}g
$$    
    
$$
  x(t) = x_0 + \frac{m}{\beta_0}v_0cos\alpha(1 - e^{-(\beta_0/m)t})
$$
    
$$
  z(t) = z_0 - \frac{m}{\beta_0}gt + \frac{m}{\beta_0}(v_0sin\alpha + g\frac{m}{\beta_0})(1-e^{-(\beta_0/m)t})
$$
</font>

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import animation
from IPython.display import HTML
import warnings 
warnings.filterwarnings('ignore')


def trajectory(t, m, v0, angle, fr=0, g=9.8, x0=0, h0=0):

    if fr == 0:
        x = x0 + v0*np.cos(angle)*t
        h = h0 + v0*np.sin(angle)*t - g*t**2/2
    else:
        x = x0 + (m/fr)*v0*np.cos(angle)*(1 - np.exp(-fr*t/m))
        h = h0 - m*g*t/fr + m/fr*(v0*np.sin(angle) + m*g/fr)*(1 - np.exp(-fr*t/m))  

    h = h[h >= 0]
    x = x[0:h.shape[0]]
    
    return [x, h]
```

Построим траектории для углов в 30, 45 и 60 градусов

```python
t = np.linspace(0, 2, 10000)
m = 100
v0 = 10

plt.figure(figsize=(20, 10))
plt.rcParams["font.size"] = "20"
plt.grid()
plt.axes().set_ylabel('Высота') 
plt.axes().set_xlabel('Длина')

angles = (np.pi/3, np.pi/4, np.pi/6)
colors = ('blue','green','red')


for alpha, color in zip(angles, colors):
    data = trajectory(t, m, v0, alpha)
    plt.plot(data[0], data[1], color=color)

plt.legend(['60 градусов','45 градусов','30 градусов'])
```

Построим траектории для различных коэффициетов сопротивления воздуха $\beta_0 = k$

```python
t = np.linspace(0, 2, 10000)
m = 100
v0 = 10

plt.figure(figsize=(20, 10))
plt.rcParams["font.size"] = "20"
plt.grid()
plt.axes().set_ylabel('Высота') 
plt.axes().set_xlabel('Длина')

friction = (0, 40, 100, 200, 400)
colors = ('blue','purple','green','red','orange')

for k, color in zip(friction, colors):
    data = trajectory(t, m, v0, np.pi/4, fr=k)
    plt.plot(data[0], data[1], color=color)
    
plt.legend(['k = 0','k = 40','k = 100','k = 200','k = 400'])
```

Построим траектории при разной гравитации

```python
plt.figure(figsize=(20, 10))
plt.grid()
plt.axes().set_ylabel('Высота') 
plt.axes().set_xlabel('Длина')

gc = (9.8, 1.62, 3.71) # Gravitational constants for Earth, Moon and Mars.
colors = ('blue','grey','red')

for gc, color in zip(gc, colors):
    data = trajectory(t, 100, 3, np.pi/6, 100, g=gc)
    plt.plot(data[0], data[1], color=color)

plt.legend(['Земля','Луна','Марс'])
```
